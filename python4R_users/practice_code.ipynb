## Methods

Python is an object-oriented language and quite different from R. 

In R, everything is a either a function or an object. In Python, **everything** is an object (e.g., list or dictionaries). Each object can do certain tasks, like be updated or altered. In stead of calling a function to an object, Python allows the object itself to call a method (e.g., head, append, etc.). 

```python
person_list = ["Jonathan", "Cornessial", "male", True, 458]
person_list
```

```python
#append values to a list
person_list.append(2018)
person_list
```

```python
#print the last element of the person_list
person_list[-1]
```

```python
person_dict = {"fname": "Jonathan", 
               "lname": "Cornessial", 
               "sex": "males", 
               "bool_value": True, 
               "twitter_followers": 458}
person_dict
```

```python
#update the person_dict dictionary with a new date & add new followers
person_dict.update({"date": "2018-06", "twitter_followers": 500})
person_dict
```

## Libraries

Vectors and matricies don't exist in the core of python BUT libraries were created for this very reason! Libraries have to be importated and dot notation can be used to access their functions. Remember to use aliases. 

```python
#import the numpy library using an alias
#import numpy as np
from sklearn import datasets
import pandas as pd

boston_data = datasets.load_boston()
df_boston = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
df_boston['target'] = pd.Series(boston_data.target)

df_boston.head(10)
```
